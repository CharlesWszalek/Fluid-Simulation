```python
# IMPLICIT FINITE SECOND-ORDER CENTRAL DIFFERENCE METHOD ON THE 1D HEAT EQUATION 
# Charles Wszalek
# 5/20/2024
# 
#     Ut = Uxx
# U(0,t) = U(1,t) = 0
# U(x,0) = x
# 
# INSPIRATION
# https://en.wikipedia.org/wiki/Finite_difference_method 
# CALCULATIONS
# https://www.overleaf.com/read/rfpqfdfwsyqt#69f6b7

# IMPORTS
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cbook, cm
from matplotlib.colors import LightSource

# import sympy as sym
```

```python
# DEFINE AXES
A = 26
xmax = 1
tmax = 1
xsteps = A
tsteps = A
x = np.linspace(0,xmax,xsteps)
t = np.linspace(0,tmax,tsteps)
X, T = np.meshgrid(x,t)
# print(X, T)
```

```python
# INITIAL CONDITIONS
Z = np.zeros_like(X*T)
for i in range(len(Z[0])):
    Z[0][i] = x[i]
Z[0][xsteps-1] = 0
# VISUALIZATION
plt.imshow(Z, cmap = 'jet')
plt.title('Heat Gradient Initial Condition')
plt.xlabel('position')
plt.ylabel('time')
plt.colorbar() 
plt.show()

plt.contourf(x, t, Z, cmap = 'jet', levels = 15) 
plt.title('Heat Gradient Initial Conditions')
plt.xlabel('position')
plt.ylabel('time')
plt.colorbar() 
plt.show()
```

```python
# MATRIX INITIALIZATION
h = xmax/xsteps
k = tmax/tsteps
r = k/(h*h)   /   4   # DIVIDING BY 4 TO SLOW THE RATE OF HEAT TRANSFER FOR SAKE OF VISUALIZATIONS
dim = (xsteps - 2) * (tsteps -1)
# for i in range(xsteps * tsteps):
#     sympy.var('u11')
M = np.zeros((dim, dim))
for i in np.arange(dim):
    for j in np.arange(dim):
        if i == j:
            M[i][j] = 1 + 2 * r
        if (i == j+1) & (np.remainder(i,xsteps-2) != 0):
            M[i][j] = -r
        if (i == j-1) & (np.remainder(j,xsteps-2) != 0):
            M[i][j] = -r
        if i == j+xsteps-2:
            M[i][j] = -1
# b = np.array([X[0][1] + r*0, X[0][2], X[0][3] + r*0, r*0, 0, r*0, r*0, 0, r*0]) # THIS IS BEFORE SIMPLIFICATION (xsteps = 5 tsteps = 4)
b = np.zeros(dim)
for i in np.arange(xsteps-2):
    b[i] = Z[0][i+1]
# print(b)
if A<=25:
    plt.imshow(M)
# np.set_printoptions(threshold=10000) # DEFAULT = 1000, USED TO SEE MORE OF MATRICES WHEN PRINTED
# print(M)
```

```python
# SOLVE FOR UNKNOWNS
a = np.linalg.solve(M, b)
# print(a)
index = 0
for i in np.arange(tsteps-1)+1:
    for j in np.arange(xsteps-2)+1:
        Z[i][j] = a[index]
        index +=1
```

```python
# 2D PLOTTING

plt.imshow(Z, cmap = 'jet')
plt.title('Heat Gradient Over Time')
plt.xlabel('position')
plt.ylabel('time')
plt.colorbar() 
plt.show()

plt.contourf(x, t, Z, cmap = 'jet', levels = 15) 
plt.title('Heat Gradient Over Time')
plt.xlabel('position')
plt.ylabel('time')
plt.colorbar()
plt.show()
```
