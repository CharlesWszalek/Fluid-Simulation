```python
# EXPLICIT FINITE DIFFERENCE METHOD ON THE HEAT EQUATION 
#     Ut = Uxx
# U(0,t) = U(1,t) = 0
# U(x,0) = x
# https://en.wikipedia.org/wiki/Finite_difference_method 

# IMPORTS
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DEFINITIONS
def isfloat(num):
    try:
        float(num)
        return 1
    except ValueError:
        return 0
```

```python
# DEFINE AXES
xmax = 1
tmax = 1
xsteps = 10
tsteps = 1000
x = np.linspace(0,xmax,xsteps)
t = np.linspace(0,tmax,tsteps)
X, T = np.meshgrid(x,t)
```

```python
# INITIAL CONDITIONS
Z = X + T
for i in range(len(Z[0])):
    Z[0][i] = x[i]
Z[0][xsteps-1] = 0
plt.imshow(Z)
# plt.contourf(x, t, Z, cmap = 'jet') 
plt.colorbar() 
plt.show()
```

```python
# PROPOGATION
h = xmax/xsteps
k = tmax/tsteps
r = k/(h*h)
for i in np.arange(len(Z[i])-2)+1: # time
    for j in np.arange(len(Z[i])-2)+1: # space
        Z[i][xsteps-1] = 0
        Z[i][0] = 0
        Z[i+1][j] = (1-2*r)*Z[i][j] + r*Z[i][j-1] + r*Z[i][j+1] 
        Z[i+1][0] = 0
        Z[i+1][xsteps-1] = 0
```

```python
# PLOTTING
plt.contourf(x, t, Z, cmap = 'jet') 
# plt.imshow(Z)
plt.colorbar() 
# plt.xlim(.8, 1)
# plt.ylim(0, 20)
plt.show()
```
