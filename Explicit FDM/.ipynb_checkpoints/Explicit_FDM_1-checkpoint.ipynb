```python
# EXPLICIT FINITE SECOND-ORDER CENTRAL DIFFERENCE METHOD ON THE 1D HEAT EQUATION # Charles Wszalek
# 5/20/2024
# 
#     Ut = Uxx
# U(0,t) = U(1,t) = 0
# U(x,0) = x
#
# INSPIRATION
# https://en.wikipedia.org/wiki/Finite_difference_method 
# CALCULATIONS
# https://www.overleaf.com/read/rfpqfdfwsyqt#69f6b7

# IMPORTS
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cbook, cm
from matplotlib.colors import LightSource
```

```python
# DEFINITIONS
def isfloat(num):
    try:
        float(num)
        return 1
    except ValueError:
        return 0
```

```python
# DEFINE AXES
xmax = 1
tmax = 1
xsteps = 25
tsteps = 25**2 * 2 
x = np.linspace(0,xmax,xsteps)
t = np.linspace(0,tmax,tsteps)
X, T = np.meshgrid(x,t)
```

```python
# INITIAL CONDITIONS
Z = X + T
for i in range(len(Z[0])):
    Z[0][i] = x[i]
Z[0][xsteps-1] = 0
'''
plt.imshow(Z)
plt.colorbar() 
plt.show()
plt.contourf(x, t, Z, cmap = 'jet') 
plt.show()
'''
```

```python
# PROPOGATION
h = xmax/xsteps
k = tmax/tsteps
r = k/(h*h)
for i in np.arange(tsteps-1): # time
    for j in np.arange(xsteps-2)+1: # space
        Z[i][xsteps-1] = 0
        Z[i][0] = 0
        Z[i+1][j] = (1-2*r)*Z[i][j] + r*Z[i][j-1] + r*Z[i][j+1] 
        Z[i+1][0] = 0
        Z[i+1][xsteps-1] = 0
```

```python
# 2D PLOTTING
# plt.imshow(Z)
# plt.colorbar() 
# plt.show()
plt.contourf(x, t, Z, cmap = 'jet') 
plt.colorbar()
plt.show()
```

```python
# 3D PLOTTING
fig, ax = plt.subplots(subplot_kw=dict(projection='3d'))
ls = LightSource(250, 30)
rgb = ls.shade(Z, cmap=cm.magma, vert_exag=0.1, blend_mode='soft')
surface = ax.plot_surface(X, T, Z, rstride=1, cstride=1, facecolors=rgb, linewidth=1, antialiased=True, shade=False)
ax.view_init(elev=40., azim=70, roll=0)
plt.show()
""" PLOT LESS OF IT
T = T[:,:round(tsteps*.2)]
Z = Z[:,:round(tsteps*.2)]
fig1, ax1 = plt.subplots(subplot_kw=dict(projection='3d'))
ls = LightSource(250, 30)
rgb = ls.shade(Z, cmap=cm.magma, vert_exag=0.1, blend_mode='soft')
surface = ax1.plot_surface(X, T, Z, rstride=1, cstride=1, facecolors=rgb, linewidth=0, antialiased=False, shade=False)
ax1.view_init(elev=40., azim=70, roll=0)
plt.show()
"""
```
